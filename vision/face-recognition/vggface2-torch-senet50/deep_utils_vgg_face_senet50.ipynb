# Download the test image

```python
import cv2
from deep_utils import show_destroy_cv2, Box, download_file, Point
from PIL import Image

# Download faces
download_file("https://raw.githubusercontent.com/pooya-mohammadi/deep-utils-notebooks/main/vision/images/face_recognition/pooya_0.jpg")
download_file("https://raw.githubusercontent.com/pooya-mohammadi/deep-utils-notebooks/main/vision/images/face_recognition/shahriar_1_0.jpg")
download_file("https://raw.githubusercontent.com/pooya-mohammadi/deep-utils-notebooks/main/vision/images/face_recognition/shahriar_2_0.jpg")

# Load an image
img_1 = cv2.imread("pooya_0.jpg")
img_2 = cv2.imread("shahriar_1_0.jpg")
img_3 = cv2.imread("shahriar_2_0.jpg")

# pil.Image is used for visualization
Image.fromarray(img_1[...,::-1]) # convert to rgb
```

# Recognize faces

```python
from scipy.spatial.distance import euclidean
# recognize the faces
result = face_recognition.extract_faces([img_1, img_2, img_3], is_rgb=False, get_time=True)


distance_1 = euclidean(result.encodings[1], result.encodings[2])
distance_2 = euclidean(result.encodings[1], result.encodings[0])
print("shariar1 and shahriar2 distance:", distance_1)
print("shariar and pooya's distance:", distance_2)
print("elaspsed time", result.elapsed_time)
```
