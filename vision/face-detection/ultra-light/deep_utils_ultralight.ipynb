## Create a face detection model using Ultralight-torch

```python
face_detector = UltralightTorchFaceDetector()
face_detector
```

# Check the input config

# Download the test image

```python
import cv2
from deep_utils import show_destroy_cv2, Box, download_file, Point
from PIL import Image

# Download an image
download_file("https://github.com/pooya-mohammadi/deep-utils-notebooks/releases/download/0.1.0/movie-stars.jpg")

# Load an image
img = cv2.imread("movie-stars.jpg")

# pil.Image is used for visualization
Image.fromarray(img[...,::-1]) # convert to rgb
```

# Detect faces

```python
# Detect the faces
result = face_detector.detect_faces(img, is_rgb=False, get_time=True)

# Draw detected boxes on the image.
img = Box.put_box(img, result['boxes'])
print(result['elapsed_time'])
# pil.Image is used for visualization
Image.fromarray(img[...,::-1]) # convert to rgb
```
